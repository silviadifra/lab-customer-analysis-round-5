```python
from sklearn.preprocessing import MinMaxScaler # do not use the function Normalise() - it does something entirely different
from sklearn.preprocessing import StandardScaler

```

```python
X = numerical[['customer lifetime value','income',
 'monthly premium auto',
 'months since last claim',
 'months since policy inception',
 'number of open complaints',
 'number of policies',
 'total claim amount',
 'months']]
y = numerical['total claim amount']

```

```python
# Normalizing data: make data range from 0 - 1, instead of from min to max
transformer = MinMaxScaler().fit(X)
X_normalized = transformer.transform(X)
display(X_normalized)

print(X_normalized.shape)

X_normalized
X_normalized_df = pd.DataFrame(X_normalized, columns=X.columns)
```

```python
sns.displot(X_normalized_df['income'])
plt.show()
```

```python
# scaling standard scaler: make data distributed with mean=0 and std=1
transformer = StandardScaler().fit(X_normalized_df)
X_normalized_standard = transformer.transform(X_normalized_df)
print(X_normalized_standard.shape)
X_normalized_standard_df = pd.DataFrame(X_normalized_standard, columns=X_normalized_df.columns)
```

```python
sns.displot(X_normalized_standard_df['income'])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_1 = X_normalized_standard_df
X_1.head()
```

```python
X_1_train, X_1_test, y_train, y_test = train_test_split(X_1, y, test_size=0.2, random_state=50)
```

```python
y_2 = pd.DataFrame(y)
y_2.info()
```

```python
print(X_1_train.shape)
print(X_1_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#we train/fit our model 
lm = LinearRegression()
lm.fit(X_1_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_train_pred = lm.predict(X_1_train)
r2_score(y_train, y_train_pred)
```

```python
y_test_pred = lm.predict(X_1_test)
r2_score(y_test, y_test_pred)
```

```python
y_test_pred
```

```python
from sklearn.metrics import mean_squared_error
# Comptuing the RMSE -> root mean squared error.
np.sqrt(mean_squared_error(y_test,y_test_pred))
```

```python
rmse=np.sqrt(mean_squared_error(y_test,y_test_pred))
rmse
```

```python
fig, ax=plt.subplots(2,2, figsize =(10,10))

sns.scatterplot(y=y_test_pred, x=y_test, ax=ax[0,0])
sns.lineplot(y=y_train, x=y_train, ax=ax[0,0])

sns.histplot(x=y_train-y_train_pred, ax=ax[0,1])
sns.scatterplot(y=y_test_pred, x=y_test, ax=ax[1,0])

sns.lineplot(y=y_test, x=y_test, ax=ax[1,0])
sns.histplot(x=y_test-y_test_pred, ax=ax[1,1])
plt.show()
```
